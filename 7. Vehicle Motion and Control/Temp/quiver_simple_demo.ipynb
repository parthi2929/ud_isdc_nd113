
# Quiver Simple Demo


A simple example of a quiver plot with a quiverkey.



```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,1,10)   # 10 numbers varying from 0 to 1
y = np.linspace(0,1,10)   # 10 numbers varying from 0 to 1
u = v = np.zeros((10,10)) # 10x10 matrix of zeroes
u[5,5] = 0.05             # one element of matrix updated with arrow size  

plt.quiver(x, y, u, v, scale=1)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = [0,0,1.225,3.675,7.35,12.25,17.15,20.8747296,21.63742781,19.07222661,14.40965867,9.886368107,7.672187372,8.82926384,12.80254538]
y = [0,0,0,0,0,0,0,-3.183769683,-8.024047717,-12.19894206,-13.7057467,-11.82164413,-7.450442506,-2.689016874,0.178566394]

u = [0]
v = [0]
for each_abs_index in range(1,len(x)):
    dx = x[each_abs_index] - x[each_abs_index-1]
    dy = y[each_abs_index] - y[each_abs_index-1]
    u.append(dx/75)  # scale it down
    v.append(dy/75)

plt.quiver(x[:-1], y[:-1], u[:-1], v[:-1], scale=1)
plt.show()
```
