### Import resources

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import numpy as np
import cv2

%matplotlib inline
```

### Read in RGB image

```python
# Read in the image
car_image = mpimg.imread('images/car_green_screen2.jpg')

plt.imshow(car_image)
```

### RGB threshold 

Visualize the green threshold you defined in the previous, consistent green color case.

```python
# Define our color selection boundaries in RGB values
lower_green = np.array([0,180,0]) 
upper_green = np.array([100,255,100])

# Define the masked area
mask = cv2.inRange(car_image, lower_green, upper_green)

# Mask the image to let the car show through
masked_image = np.copy(car_image)

masked_image[mask != 0] = [0, 0, 0]

# Display it!
plt.imshow(masked_image)
```

### Convert to HSV

```python
# Convert to HSV
car_image_hsv = cv2.cvtColor(car_image, cv2.COLOR_RGB2HSV)

# HSV channels
h = car_image_hsv[:,:,0]
s = car_image_hsv[:,:,1]
v = car_image_hsv[:,:,2]

# Visualize the individual color channels
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20,10))
ax1.set_title('H channel')
ax1.imshow(h, cmap='gray')
ax2.set_title('S channel')
ax2.imshow(s, cmap='gray')
ax3.set_title('V channel')
ax3.imshow(v, cmap='gray')
```
